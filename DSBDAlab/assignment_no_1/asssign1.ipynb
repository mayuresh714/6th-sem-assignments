## NaN values

```python
def get_different(col,df):
    diction = {}
    for val in df.index:
        cell_val = df.loc[val,col]
        try:
            diction[type(cell_val)] += 1
        except:
            diction[type(cell_val)] = 1
    return diction
```

```python
for i in dfnew.columns:
    value = get_different(i,dfnew)
    print(i , "         ",value,"          ",len(value)   )
```

```python
of.Type.value_counts()
```

```python
def map(x):
    dict_ = {'h':1,"u":2,"t":3}
    return dict_[x] 
```

```python
of.Type = of.Type.apply(lambda x: map(x))
```
